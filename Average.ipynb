## Calculating RGB values and then calculating average

```python
import cv2
import glob
```

```python
sumr = 0
sumg = 0
sumb = 0
count = 0
sr = 0
sg = 0
sb = 0
path=glob.glob(r"G:\Dr. Kale\Healthy Flash 16 images resized\H 021 Trushna Baramse 36 Flash\*.jpg")
print("Red        Green      Blue")
for file in path:
    img = cv2.imread(file)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    # considering 10 pixels from each rows of 5
    for i in range(0, 256, 52):
        for j in range(0, 256, 27):
            color = img[i, j]
            red = color[0]
            green = color[1]
            blue = color[2]
            sumr += red
            sumg += green
            sumb += blue
            print(f"{red}   {green}   {blue}")
            r = sumr/50
            g = sumg/50
            b = sumb/50
    count += 1
    # average of RGB values of each image of blood samples
    print(f"{r}      {g}      {b}")
    sr += r
    sg += g
    sb += b
    sumr = sumg =sumb =0
print(f"{count}")
print(f"{sr}  {sg}  {sb}")
ar = sr/count
ag = sg/count
ab = sb/count
# average of average 
print(f"{round(ar, 5)}   {round(ag, 5)}   {round(ab, 5)}")
print(f"Number of images of a patient:{count}")
cv2.waitKey(0)
```
